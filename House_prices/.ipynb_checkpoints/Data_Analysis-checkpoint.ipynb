### **Null 데이터 처리**

```python
#Find Null data in train and test data
na_train = train_df.isna().sum()
print("train data null값 존재 feature와 NaN 개수")
display(na_train[na_train >=1])
print("train data null feature 개수 :", len(na_train[na_train >= 1].index.tolist()))
print("-"*50)
na_test = test_df.isna().sum()
print("test data null값 존재 feature와 NaN 개수")
display(na_test[na_test >=1])
print("train data null feature 개수 :", len(na_test[na_test >= 1].index.tolist()))
```

```python
#Delete features which have over 500 NaN values
drop_na_features = na_train[na_train > 500].index.tolist()
train_df.drop(drop_na_features, axis = 1, inplace = True)
test_df.drop(drop_na_features, axis = 1, inplace = True)
```

```python
na_train = train_df.isna().sum()
na_test = test_df.isna().sum()

#Find common ground between both
common_list = []
for ele in na_test[na_test>=1].index.tolist():
    if ele in na_train[na_train>=1].index.tolist():
        common_list.append(ele)
print("train과 test가 공통적으로 가지고 있는 Null 보유 feature들 :", common_list)
print("총 개수 :", len(common_list))
print("-"*150)

#Find NaN feature only in train
train_NaOnly = []
for ele in na_train[na_train>=1].index.tolist():
    if ele not in common_list:
        train_NaOnly.append(ele)
print("train만 가지고 있는 Null 보유 feature들 :", train_NaOnly)
print("총 개수 :", len(train_NaOnly))
print("-"*150)

#Find NaN feature only in test
test_NaOnly = []
for ele in na_test[na_test>=1].index.tolist():
    if ele not in common_list:
        test_NaOnly.append(ele)
print("test만 가지고 있는 Null 보유 feature들 :", test_NaOnly)
print("총 개수 :", len(test_NaOnly))
```

**적은 수(1~10)의 Missig Value에 대한 처리**

1. 모델 학습 알고리즘에 맡긴다 (예, xgboost와 같은 모델은 자체적으로 missing value 처리에 효과적)

**(11~100)의 Missing Value에 대한 처리**

위와 같이 OHE 적용 유무로 나누자

해당되는 feature들

1. train - BsmtExposure, GarageType, GarageFinish

**Category**

MSZoning / Street / LotShape / LandContour / Utilities / LotConfig / LandSlope / Neighborhood / Condition1 / Condition2 / BldgType / HouseStyle / RoofStyle / RoofMatl / Exterior1st / Exterior2nd / Foundation / BsmtExposure / Heating / CentralAir / Electrical / Functional / GarageType / GarageFinish / PavedDrive / SaleType / SaleCondition

-> OHE 적용

**Rank**
ExterQual / ExterCond / BsmtQual / BsmtCond / BsmtFinType1 / BsmtFinType2 / HeatingQC / KitchenQual / GarageQual / GarageCond

-> Label Encoding 적용

```python
OHE_list = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities', 'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 
            'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'Foundation', 'BsmtExposure', 'Heating', 'CentralAir',
            'Electrical', 'Functional', 'GarageType', 'GarageFinish', 'PavedDrive', 'SaleType', 'SaleCondition']
Rank_list = ['OverallQual', 'OverallCond', 'ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'BsmtFinType1', 'BsmtFinType2', 
             'HeatingQC', 'KitchenQual', 'GarageQual', 'GarageCond']
```

```python
#Find some values that exist in test data but not in train - error
error_list = []
for ele in OHE_list:
    train_list = train_df[ele].values.tolist()
    test_list = test_df[ele].values.tolist()
    for ele2 in test_list:
        if ele2 not in train_list:
            error_list.append(ele)

print("문제 있는 category형 column 개수 :", len(error_list))
print(error_list)

error_list = []
for ele in Rank_list:
    train_list = train_df[ele].values.tolist()
    test_list = test_df[ele].values.tolist()
    for ele2 in test_list:
        if ele2 not in train_list:
            error_list.append(ele)

print("문제 있는 rank형 column 개수 :", len(error_list))
print(error_list)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

#Do OHE on category feature
for ele in OHE_list:
    ohe = OneHotEncoder(sparse_output = False)
    train_ohe = ohe.fit_transform(train_df[ele].values.reshape(-1,1))
    test_ohe = ohe.transform(test_df[ele].values.reshape(-1,1))
    train_df = pd.concat([train_df, pd.DataFrame(train_ohe, columns = [ele + '_' + i for i in ohe.categories_[0]])], axis = 1)
    test_df = pd.concat([test_df, pd.DataFrame(test_ohe, columns = [ele + '_' + i for i in ohe.categories_[0]])], axis = 1)

train_df.drop(OHE_list, axis = 1, inplace = True)
test_df.drop(OHE_list, axis = 1, inplace = True)


#Do LaebelEncoding on rank feature
for ele in Rank_list:
    ohe = LabelEncoder()
    train_ohe = ohe.fit_transform(train_df[ele])
    test_ohe = ohe.transform(test_df[ele])
    train_df = pd.concat([train_df, pd.DataFrame(train_ohe, columns = [ele + '_label'])], axis = 1)
    test_df = pd.concat([test_df, pd.DataFrame(test_ohe, columns = [ele + '_label'])], axis = 1)

train_df.drop(Rank_list, axis = 1, inplace = True)
test_df.drop(Rank_list, axis = 1, inplace = True)
```

### **Scaling**

log conversion 적용 가정

```python
log_list = ['LotFrontage', 'LotArea', 'MasVnrArea', 'TotalBsmtSF', 'LowQualFinSF', 'GrLivArea', 'WoodDeckSF', 'OpenPorchSF', 
            'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal']
```

```python
#Apply log conversion
log_tr = train_df.copy(deep = True)
log_te = train_df.copy(deep = True)

for ele in log_list:
    log_tr[ele] = np.log1p(log_tr[ele])

for ele in log_list:
    log_te[ele] = np.log1p(log_te[ele])

scalecheck_trdf = log_tr[['LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtUnfSF', 
                'TotalBsmtSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 
                'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 
                'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']]

scalecheck_tedf = log_te[['LotFrontage', 'LotArea', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtUnfSF', 
                'TotalBsmtSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 
                'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 
                'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']]
```

```python
# Check max/min of features
for ele in scalecheck_trdf:
    print(ele,"의 max / min :",  log_tr[ele].max(),"/", log_tr[ele].min())
    print("-"*150)
```
